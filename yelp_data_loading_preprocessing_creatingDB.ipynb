```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dask import dataframe as dd
import time
import ast
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
Path = '/content/drive/MyDrive/dataset'

csv = ['business.csv','review.csv','tip.csv','user.csv']
```

```python
def csv_read(file_no):
  s_time_d = time.time()
  ddf = dd.read_csv(Path+'/'+csv[file_no])
  print('No.of partitions taken to convert to dataframe while performing parallel:',ddf.npartitions)
  e_time_d = time.time()
  print("time taken to read",e_time_d - s_time_d)
  ddf = ddf.compute()
  print(ddf.shape)
  return ddf
```

```python
userid_to_idx = {user : idx for idx, user in enumerate(user.user_id.unique())}
businessid_to_idx = {b_id : idx for idx, b_id in enumerate(bus.business_id.unique())}
```

```python
def idx_convertion(df,u = False ,b = False):
  if u:
    df.user_id = df.user_id.map(lambda x: userid_to_idx[x])
  if b:
    df.business_id = df.business_id.map(lambda x: businessid_to_idx[x])
    pass
```

```python
bus = csv_read(0)
bus.head()
```

```python
filter = bus.categories.str.contains('Restaurant', na=False) | bus.categories.str.contains('Food', na=False)
bus_f = bus[filter]
```

```python
bus_f.isnull().sum()
```

```python
bus_f.fillna(False, inplace = True)
```

```python
val = ast.literal_eval(bus_f['attributes'].iloc[0])
cols = list(val.keys())

def add_cols(name):
  cols.remove(name)
  for i in list(ast.literal_eval(val[name]).keys()):
    cols.append(name+'_'+i)
add_cols('BusinessParking')
add_cols('Ambience')
add_cols('GoodForMeal')

cols_days = list(ast.literal_eval(bus_f['hours'].iloc[0]))
cols.extend(cols_days)
cols
```

```python
def dict_process(var1,var2,d,i):
  if var1 in cols:
    if isinstance(d[var2], bool) :
      att_df[var1].loc[i]=d[var2]
    elif (d[var2] == 'None') or (d[var2] == None) :
      att_df[var1].loc[i]=False
    else:
      att_df[var1].loc[i]=d[var2].lstrip('u')

def value_process(val,i):
  if val != False:
    d = ast.literal_eval(val)
    print(d)
    for j in d.keys():
      dict_process(j,j,d,i)
      if j in ['Ambience','BusinessParking','GoodForMeal']:
        d1 = ast.literal_eval(d[j])
        print(d1,d[j])
        try:
          for k in d1.keys():
            dict_process(j+'_'+k,k,d1,i)
        except:
          pass
```

```python

att_df = pd.DataFrame(False , index = bus_f.index,columns = cols)
for i in range(bus_f.shape[0]):
  value = bus_f['attributes'].iloc[i]
  value_days = bus_f['hours'].iloc[i]
  value_process(value,i)
  value_process(value_days,i)
  
att_df
```

```python
bus_updated = pd.concat([bus_f, att_df], axis='columns')
```
